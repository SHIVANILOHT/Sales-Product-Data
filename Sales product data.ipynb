```python
df_jan = pd.read_csv("Sales_January_2019.csv")
df_feb = pd.read_csv("Sales_February_2019.csv")
df_mar = pd.read_csv("Sales_March_2019.csv")
df_apr = pd.read_csv("Sales_April_2019.csv")
df_may = pd.read_csv("Sales_May_2019.csv")
df_jun = pd.read_csv("Sales_June_2019.csv")
df_jul = pd.read_csv("Sales_July_2019.csv")
df_aug = pd.read_csv("Sales_August_2019.csv")
df_sep = pd.read_csv("Sales_September_2019.csv")
df_oct = pd.read_csv("Sales_October_2019.csv")
df_nov = pd.read_csv("Sales_November_2019.csv")
df_dec = pd.read_csv("Sales_December_2019.csv")

df_sales = pd.concat([df_jan, df_feb, df_mar, df_apr, df_may, df_jun, df_jul, df_aug , df_sep , df_oct, df_nov, df_dec])
df_sales  = df_sales[df_sales["Order ID"].notnull()]
df_sales = df_sales[df_sales["Order ID"]!="Order ID"]
df_sales['Price Each'] = df_sales['Price Each'].apply(float)
df_sales["Order Date_"] = df_sales["Order Date"].str.split(" ",expand =True)[0]
df_sales['Order Date_'] = pd.to_datetime(df_sales["Order Date_"],infer_datetime_format = True)
df_sales["year"] = df_sales["Order Date_"].dt.year
df_sales ["Month"] = df_sales["Order Date_"].dt.month
df_sales['city']=[x.split(',')[1] for x in df_sales['Purchase Address'].values]
```
